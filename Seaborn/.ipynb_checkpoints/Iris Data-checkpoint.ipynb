```python
import pandas as pd
import seaborn as sns
import numpy as np
```

# The Iris dataset is provided by deafult by the Seaborn module

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
species = iris['species']
species.value_counts()
```

```python
sns.scatterplot(y = 'petal_length', x = 'sepal_length', data = iris)
```

```python
sns.scatterplot(y = 'petal_length', x = 'sepal_length', hue = 'species', data = iris)
```

```python
sns.scatterplot(y = 'petal_length', x = 'sepal_length', data = iris, hue = 'species', size = 'species')
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue = 'species', palette = 'Set1')
```

```python
sns.lmplot(x = 'petal_length', y = 'petal_width', data = iris)
```

```python
 sns.lmplot(x = 'petal_length', y = 'petal_width', data = iris, hue = 'species', markers = ['o', 'x', '>']) 
```

```python
sns.boxplot(x = 'species', y = 'sepal_width', data = iris)
```
